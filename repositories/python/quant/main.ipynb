```python
import json
from datetime import datetime
import pytz

local_tz = pytz.timezone("Asia/Shanghai")  # 替换为你的本地时区



# 假设 JSON 文件为 data.json
file_path = "007-20241119.json"

# 打开并读取文件内容
with open(file_path, "r", encoding="utf-8") as file:
    data = json.load(file)

#print(data)  # 输出 JSON 数据，类型为 Python 字典或列表

arr = []
ma = []
sum = 0
cnt = 0
time = []
dist = []

for item in data:
  cnt+=1
  sum += item["price"]
  arr.append(item["price"])
  ma.append(sum/cnt)
  time.append(datetime.fromtimestamp(item["time"],local_tz))
  dist.append((item["price"]/(sum/cnt)-1)*100)

for i in range(len(arr)):
  print(datetime.fromtimestamp(data[i]["time"],local_tz),arr[i], int(ma[i]), (int(arr[i])/ma[i] - 1)*100)
import matplotlib.pyplot as plt
# plt.plot(time, arr)
# plt.plot(time, ma)

plt.plot(time, dist)
plt.gcf().autofmt_xdate()
plt.show()

```
